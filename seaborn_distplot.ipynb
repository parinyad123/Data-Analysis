```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_formate='retina' # ความละเอียดระดับ retina
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

# การสร้างกราฟ Histogram ด้วย distplot

```python
sns.distplot(df['cty'])
```

```python
sns.distplot(df['cty'], kde=False) #  ไม่ต้องการเส้น kde
plt.axvline(x=df.cty.mean(), color="orange", linestyle="--") # เพิ่มเส้นค่าเฉลี่ย
plt.title(f"mean = {df.cty.mean():.2f}, sd ={df.cty.std():.2f}") # ใส่รายละเอียด (:.2f ปรับเป็นทศนิยม 2 ตำแหน่ง)
```

#### จะเห็นได้ว่า histogram ปกติที่ไม่มี kde แกน y จะเป็นจำนวนรุ่น (cty) แต่ถ้ามี kde แกน y จะเป็น density

### ทำการเปรียบเทียบ cty กับ hwy

```python
sns.distplot(df['cty'], kde=False)
sns.distplot(df['hwy'], kde=False)
```

```python
sns.distplot(df['cty'], kde=False, label = 'city')
sns.distplot(df['hwy'], kde=False, label = 'highway')
plt.legend()
```

```python
sns.distplot(df['cty'], kde=True, label = 'city')
sns.distplot(df['hwy'], kde=True, label = 'highway')
plt.legend()
```

```python
sns.distplot(df['cty'], kde=False, label = 'city')
plt.show()
sns.distplot(df['hwy'], kde=False, label = 'highway')
plt.show()
```

```python
# สร้างกราฟ 2 กราฟ
fig,ax=plt.subplots(1,2) # subplot แบบ row 1 column 2 
print(ax.shape)
```

```python
# จากนั้นเอาภาพมาใส่ในกราฟ
fig,ax=plt.subplots(1,2) # subplot แบบ row 1 column 2 
print(ax.shape)
sns.distplot(df['cty'], kde=True, ax=ax[0]) # ax[ตำแหน่งภาพ]
sns.distplot(df['hwy'], kde=True, ax=ax[1])
```

```python
# จากนั้นเอาภาพมาใส่ในกราฟ
fig,ax=plt.subplots(1,2,sharex=True, sharey=True ) # subplot แบบ row 1 column 2 , sharey=True ใช้แชร์แกน y 
print(ax.shape)
sns.distplot(df['cty'], kde=False, ax=ax[0]) # ax[ตำแหน่งภาพ] ไม่เอา kde เพราะอยากดูความถี่
sns.distplot(df['hwy'], kde=False, ax=ax[1])
```

```python
# เปลี่ยนสีกราฟ
# p = ['maroon','lightgray','maroon','deepskyblue','maroon','maroon','maroon']

p = ["deepskyblue" if v > 40 else 'lightgray' 
     for v in df['class'].value_counts()] # กำหนดสีตามค่า ถ้า > 40 เป็นสีฟ้า ถ้าไม่สีเทา

# p = ["deepskyblue" if v == 'pickup' else "lightgray" 
#     for v in df['class'].value_counts().index]  # highlight เฉพาะกราฟ





sns.countplot(x='class', data=df, 
              palette=p,
              order =df['class'].value_counts().index)
```

### add horizontal line

```python
avg = df['class'].value_counts().mean() # คำนวณค่าเฉลี่ย

sns.countplot(x = 'class', data= df, color = 'lightgray',
             order = df['class'].value_counts().index)
plt.axhline(y=avg, linestyle = '--', color = 'deepskyblue') # เพิ่มเส้น horizontal
```

### add vertical line

```python
avg = df['class'].value_counts().mean() # คำนวณค่าเฉลี่ย

sns.countplot(y = 'class', data= df, color = 'lightgray',
             order = df["class"].value_counts().index)
plt.axvhline(x=avg, linestyle='--', color = 'deepskyblue') # เพิ่มเส้น horizontal
# plt.ylabel=''
```

# การใช้ factorplot เพื่อสร้างกราฟหลายๆ รูปตาม foctor ที่กำหนด

```python
df.head()
```

### ทำการ clean data
#### โดยการดึงอักษรตัวแรกของ trans มา ถ้าเป็น a ให้เปลี่ยนเป็น auto ถ้า m เป้น manual

```python
df['gear']=df.trans.str[0].map({'a':'auto','m':'manual'}).astype("category")
df['drv']=df.drv.map({'f':'front','r':'rear','4':'4-wheel'}).astype("category")
df.head()
```

```python
sns.barplot(x='year', y='cty', data=df, ci=None)
```

### add hue

```python
sns.barplot(x='year', y='cty', data=df,
           hue='gear', ci=None)
```

### use factorplot

```python
sns.barplot(x='year', y='cty', data=df,
           hue ='gear',
           size =4 , aspect=1)
```

# การสร้างกราฟหลายรูปด้วย FacetGrid

```python
sns.distplot(df['cty'])
```

```python
g = sns.FacetGrid(df, col='drv') # col=column
g.map(sns.distplot, 'cty')
```

```python
g = sns.FacetGrid(df, col='drv') # col=column
g.map(sns.distplot, 'cty', kde=False, color='.75')
```

```python
g = sns.FacetGrid(df, col='drv', size = 4, aspect=1) # size กำหนดแกน x , aspect กำหนดขนาดแกน y โดย 1 x size
g.map(sns.distplot, 'cty', kde=False, color = '.25')
```

```python
g = sns.FacetGrid(df, row='gear', col='drv')
g.map(sns.distplot,'cty')
```

```python
g = sns.FacetGrid(df, row='gear', col='drv')
g.map(sns.kdeplot,'cty', shade=True, color ='tan') # plot kde
```

```python
g = sns.FacetGrid(df, row='gear', col='drv', hue='year')
g.map(sns.kdeplot,'cty')
```

```python
# use native matplotlib plt.scatter
g = sns.FacetGrid(df, row='gear', col='drv')
g.map(plt.scatter, 'displ','cty', alpha=.5, edgecolor='white')
```

```python
g = sns.FacetGrid(df, row='gear', col='drv')
# g.map(sns.regplot, x='displ', y='cty', data=df, fit_reg=False)
# g.map(sns.regplot, x='displ', y='cty', data=df, fit_reg=True, scatter_kws={'alpha':3})
g.map(sns.regplot, x='displ', y='cty', data=df)
```

```python
g = sns.FacetGrid(df, row='gear', col='drv')
# g.map(sns.regplot, x='displ', y='cty', data=df, fit_reg=False)
g.map(sns.barplot, x='drv', y='cty', data=df, palette='Set2', order=['front', 'rear', '4-wheel'])
```

```python
g = sns.FacetGrid(df, row='gear', col='year')
# g.map(sns.regplot, x='displ', y='cty', data=df, fit_reg=False)
g.map(sns.boxplot, x='drv', y='cty', data=df, palette='Set2', order=['front', 'rear', '4-wheel'])
```
